```python
df['fecha_registro'] = pd.to_datetime(
    df['fecha_registro'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)
```

```python
df['gasto_q'] = df['gasto_q'].str.replace(',', '.', regex=False)
```
